```python
from sklearn.preprocessing import Imputer

def load_data():
    print("Load the data.")
    
    global y_train
    global x_train
    global x_test
    df_x_train = pd.concat(map(pd.read_csv, train_feature_files),axis=0)
    df_x_test = pd.concat(map(pd.read_csv, test_feature_files),axis=0)
    df_y_train = pd.read_csv("train_labels.csv")

    x_test = df_x_test.values
    x_train = df_x_train.values
    y_train = df_y_train['y'].values

    print('Standardize the data.')
    scaler = preprocessing.StandardScaler().fit(x_train)
    x_train =scaler.transform(x_train)
    x_test = scaler.transform(x_test)
    
    print('Impute missing data.')
    imputer = Imputer(missing_values='NaN', strategy='mean', axis=0)
    imputer = imputer.fit(x_train)
    x_train = imputer.transform(x_train)
    x_test = imputer.fit_transform(x_test)
    
```

```python
from sklearn.svm import SVC
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import balanced_accuracy_score
# F1 = f1_score(y_true, y_pred, average='micro')
load_data()
n=3

Cvalues = [0.1,0.3,0.5,1]
# Cvalues = [5,5.2,5.4,5.6,5.8,6]

def train(C):
    val_result = []

    for i in range(n):
        x_val_t = x_train[np.arange(21600*i,21600*(i+1)),:]
        y_val_t = y_train[np.arange(21600*i,21600*(i+1))]
        mask = np.ones(x_train.shape[0],dtype=bool)
        mask[np.arange(21600*i,21600*(i+1))] = 0
        x_train_t = x_train[mask,:]
        y_train_t = y_train[mask]
        
        np.random.seed(i)
#         x_train_t, y_train_t = outlier_detection(x_train,y_train,method='isoforest')
        
#         model_RandomUnderSampler = RandomUnderSampler()
#         x_train_t, y_train_t =model_RandomUnderSampler.fit_sample(x_train_t,y_train_t)
#         print('After down-sampling:\n',pd.DataFrame(y_train_t,columns=['y']).groupby('y').size())

#         x_train_t, x_val_t, y_train_t, y_val_t = train_test_split(x_train, y_train, test_size=0.1)
#         x_train_t, x_val_t, y_train_t, y_val_t = train_test_split(x_train_t, y_train_t, test_size=0.1)

        clf = SVC(kernel='rbf',C=C, 
        decision_function_shape='ovo',  #gamma='auto',
        shrinking=True,
        verbose=False,class_weight="balanced")   # add class weight

        clf.fit(x_train_t, y_train_t)

        y_val_p = clf.predict(x_val_t)
        val_result.append(balanced_accuracy_score(y_val_t, y_val_p))

        y_pred = clf.predict(x_test)
        print("predicted y:",y_pred)


        f = open("submission_svc{0:.1f}.csv".format(i), "w")
        f.write("id,y\n")
        for i,x in enumerate(y_pred):
            f.write("{},{}\n".format(i,x))
        f.close()

    print("C value: %f\nCross validation score:%f"%(C,np.mean(val_result)))

# for C in Cvalues:
#     train(C)
train(0.1)
```

```python
size = x_test.shape
result = [list() for i in range(size[0])]


for i in range(n):
    f = "submission_svc%d.0.csv" %(i)
    f = open(f)
    lines = f.readlines()
    for l in lines[1:]:
        l = l.strip().split(',')
        idx, val = int(l[0]), int(float(l[1]))
        result[idx].append(val)
    f.close()
# print(result)


def vote(x):
    c = [0] * 4
    for i in x:
        c[i] += 1
    c = [(1, c[1]), (2, c[2]), (3, c[3])]
    c.sort(key = lambda x: x[1], reverse = True)
    if c[0][1] > c[1][1]:
        return c[0][0]
    else:
        if c[0][0] == 1 or c[1][0] == 1:
            return 1
        else:
            return np.random.choice([c[0][0], c[1][0]])

print(vote([1,2,3,3,3]))


with open("voted_svc.csv", "w") as f:
    f.write("Id,y\n")
    for i in range(size[0]):
        f.write("{},{}\n".format(i, vote(result[i])))
    f.close()
```
